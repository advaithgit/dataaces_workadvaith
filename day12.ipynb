```python
import pandas as pd
import spacy

nlp = spacy.load('xx_ent_wiki_sm')

input_file_path = 'preprocessed.xlsx'
df = pd.read_excel(input_file_path)

def extract_unique_entities(doc):
    entities = {}
    for ent in doc.ents:
        label = ent.label_
        text = ent.text
        if label not in entities:
            entities[label] = []
        entities[label].append(text)
    return entities

for index, row in df.iterrows():
    text = row['title'] + ' ' + row['snippet'] + ' ' + row['content']
    doc = nlp(text)

    entities = extract_unique_entities(doc)

    for label, entity_list in entities.items():
        df.at[index, label] = '/ '.join(entity_list)

output_file_path = 'FINAL_WIKI.xlsx'
df.to_excel(output_file_path, index=False)
```

```python
import pandas as pd
import spacy
from collections import Counter

nlp = spacy.load('en_core_web_sm')


input_file_path = 'preprocessed.xlsx'
df = pd.read_excel(input_file_path)


def extract_top_entities(doc):
    entities = {}
    for ent in doc.ents:
        label = ent.label_
        text = ent.text
        if label not in entities:
            entities[label] = Counter()
        entities[label][text] += 1

    top_entities = {}
    for label, entity_counts in entities.items():
        top_entities[label] = [text for text, _ in entity_counts.most_common(2)]

    return top_entities

for index, row in df.iterrows():
    text = row['title'] + ' ' + row['snippet'] + ' ' + row['content']
    doc = nlp(text)

    top_entities = extract_top_entities(doc)

    for label, entity_list in top_entities.items():
        df.at[index, label] = '/ '.join(entity_list)


output_file_path = 'outputfile_with_top_entities.xlsx'
df.to_excel(output_file_path, index=False)
```

```python
import pandas as pd
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Read the input Excel file
input_file_path = 'preprocessed.xlsx'
df = pd.read_excel(input_file_path)

# Define a function to extract the top 2 entities for each label in a spaCy Doc using TF-IDF
def extract_top_entities(doc):
    entities = {}
    for ent in doc.ents:
        label = ent.label_
        text = ent.text
        if label not in entities:
            entities[label] = Counter()
        entities[label][text] += 1

    top_entities = {}
    for label, entity_counts in entities.items():
        # Combine entity texts into a single string for TF-IDF analysis
        entity_text = [' '.join([text] * count) for text, count in entity_counts.items()]

        # Skip if entity_text is empty or contains very short text
        if not entity_text or all(len(text.split()) <= 1 for text in entity_text):
            continue

        vectorizer = TfidfVectorizer()
        print(entity_text)
        tfidf_matrix = vectorizer.fit_transform(entity_text)

        # Check if TF-IDF matrix is empty
        if tfidf_matrix.shape[1] == 0:
            continue

        # Find the top 2 entities based on TF-IDF scores
        top_indices = tfidf_matrix.sum(axis=0).argsort()[0, -2:]
        top_entities[label] = [vectorizer.get_feature_names_out()[index] for index in top_indices]

    return top_entities

# Process each row in the DataFrame
for index, row in df.iterrows():
    text = row['title'] + ' ' + row['snippet'] + ' ' + row['content']
    doc = nlp(text)

    top_entities = extract_top_entities(doc)

    for label, entity_list in top_entities.items():
        # Ensure that all elements in entity_list are converted to strings
        entity_list = [str(entity) for entity in entity_list]
        df.at[index, label] = '/ '.join(entity_list)

# Save the modified DataFrame to a new Excel file
output_file_path = 'outputfile_with_top_entities_tfidf.xlsx'
df.to_excel(output_file_path, index=False)
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)

input_file_path = 'preprocessed.xlsx'
df = pd.read_excel(input_file_path)

for label in nlp.model.config.id2label.values():
    df[label] = ""

for index, row in df.iterrows():
    text = row['title'] + ' ' + row['snippet'] + ' ' + row['content']

    print(text)
    ner_results = nlp(text)

    for result in ner_results:
        label = result['entity']
        entity = result['word']
        df.at[index, label] += entity + ', '

for label in nlp.model.config.id2label.values():
    df[label] = df[label].str.rstrip(', ')

output_file_path = 'outputfile1_bert_ner.xlsx'
df.to_excel(output_file_path, index=False)
```
