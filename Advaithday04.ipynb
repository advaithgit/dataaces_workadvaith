```
import pandas as pd
from transformers import pipeline

# Load the dataset
excel_file_path = 'Dataset2.xlsx'  # Replace with your Excel file path
df = pd.read_excel(excel_file_path)

# Specify the column name containing the content
content_column_name = 'content'

# Initialize transformer for summarization
summarizer = pipeline(task="summarization", model="facebook/bart-large-xsum", tokenizer="facebook/bart-large-cnn")

#Inititalise summary variable
summaries=[]
# Iterate through each row and print summaries
for index, row in df.iterrows():
    content = row[content_column_name]
    try:
        summary = summarizer(content, max_length=150, min_length=30, do_sample=True)[0]['summary_text']
        summaries.append(summary)
        print(f"Summary for Row {index + 1}:\n{summary}\n")
    except Exception as e:
        print(f"Error generating summary for row {index + 1}: {e}")
        summaries.append(content)
df['summaries'] = summaries

# Save the updated DataFrame to a new Excel file
updated_excel_file_path = 'UpdatedExcelFile.xlsx'
df.to_excel(updated_excel_file_path, index=False)

print("Summaries added and new Excel file created.")

```
